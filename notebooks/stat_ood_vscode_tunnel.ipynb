## 3. Run the Tunnel
Run the cell below. It will output a code and a URL (e.g., `https://github.com/login/device`).

1.  Click the link and enter the code to authenticate with your GitHub account.

3.  **DO NOT** open that link in the browser if you want to use Desktop VS Code.
4.  Instead, go to your **Local VS Code**:
    *   Install the **"Remote - Tunnels"** extension.
    *   Click the Remote Explorer icon (monitor with ><).
    *   Select **Tunnels** from the dropdown.
    *   You should see this Colab instance listed (e.g., `colab-machine`).
    *   Click the arrow `->` to connect.
